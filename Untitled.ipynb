```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
ax.grid(axis="y",linestyle=":",zorder=0)

t_tx_avg = 0.294
t_tx_max = 0.338
t_tx_min = 0.222
t_tx_var = 0.116
t_tx_mdev = 0.018

p_tx_avg = 0.177
p_tx_max = 0.210
p_tx_min = 0.94
p_tx_var = 0.109
p_tx_mdev = 0.014

objects = ("Traditional","Proposed")
y_pos = np.arange(len(objects))
latency_avg = [t_tx_avg,p_tx_avg]
e_bar = [t_tx_mdev,p_tx_mdev]
e_bar_set = dict(lw = 1, capthick = 1, capsize = 10)

ax.bar(y_pos,latency_avg,align="center",width=0.6,linewidth=3,color="#61C1BE",edgecolor="#00ADA9",yerr=e_bar,error_kw=e_bar_set,zorder=3)
plt.xticks(y_pos,objects,size = 14)
ax.set_ylim([0.00,0.35])
ax.set_ylabel("latency(usec)",size=14)
ax.set_title("Transmit Latency",size=14)

ax.legend()
plt.savefig("tx_latency.png", format="png", dpi=300)
plt.show()
```
